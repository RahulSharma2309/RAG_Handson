```python
from pathlib import Path
import json
import hashlib
import pandas as pd
import tiktoken

print('Embeddings handson imports loaded')
```

```python
# Paths
BASE = Path('../data').resolve()
CHUNK_INPUT_DIR = BASE / 'chunk_inputs'
EMBED_ARTIFACT_DIR = BASE / 'embedding_artifacts'
QUALITY_DIR = BASE / 'quality_reports'

for p in [CHUNK_INPUT_DIR, EMBED_ARTIFACT_DIR, QUALITY_DIR]:
    p.mkdir(parents=True, exist_ok=True)

print('chunk_inputs      :', CHUNK_INPUT_DIR)
print('embedding_artifacts:', EMBED_ARTIFACT_DIR)
print('quality_reports    :', QUALITY_DIR)
```

```python
def count_tokens(text: str, model_name: str = 'gpt-4o-mini') -> int:
    enc = tiktoken.encoding_for_model(model_name)
    return len(enc.encode(text))

def stable_embedding_id(text: str, source: str, chunk_index: int) -> str:
    base = f'{source}|{chunk_index}|{text}'
    return 'emb_' + hashlib.sha1(base.encode('utf-8')).hexdigest()[:16]

print('Utility methods ready')
```
