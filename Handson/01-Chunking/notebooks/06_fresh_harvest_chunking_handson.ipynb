# 06 - Fresh Harvest Documents Chunking Hands-On

This notebook chunks only `Documents/Fresh_Harvest_Documents` with a technical-doc profile.

Outputs:
- per-file chunk artifacts: `data/fresh_harvest_documents/chunk_artifacts`
- quality reports: `data/fresh_harvest_documents/quality_reports`


````python
from pathlib import Path
from datetime import datetime
import hashlib, json, re
import pandas as pd
import tiktoken
from langchain_core.documents import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import TextLoader

enc = tiktoken.get_encoding('cl100k_base')
print('Environment ready')
````

````python
REPO_ROOT = Path(r"C:/Users/Lenovo/source/repos/RAG_Handson")
SRC_ROOT = REPO_ROOT / "Documents" / "Fresh_Harvest_Documents"
OUT_CHUNKS = REPO_ROOT / "Handson" / "01-Chunking" / "data" / "fresh_harvest_documents" / "chunk_artifacts"
OUT_REPORTS = REPO_ROOT / "Handson" / "01-Chunking" / "data" / "fresh_harvest_documents" / "quality_reports"
OUT_CHUNKS.mkdir(parents=True, exist_ok=True)
OUT_REPORTS.mkdir(parents=True, exist_ok=True)

ALLOWED_EXT = {'.md', '.mmd', '.txt', '.ps1', '.py', '.csv'}
files = sorted([p for p in SRC_ROOT.rglob('*') if p.is_file() and p.suffix.lower() in ALLOWED_EXT])
print('Files discovered:', len(files))
for f in files[:10]:
    print('-', f.relative_to(REPO_ROOT))
````

````python
def count_tokens(text: str) -> int:
    return len(enc.encode(text))

def chunk_id(text: str) -> str:
    return 'fh_' + hashlib.md5(text.encode('utf-8')).hexdigest()[:12]

def clean_text(text: str) -> str:
    text = text.replace('\x00', '').replace('ﬁ','fi').replace('ﬂ','fl')
    text = re.sub(r'\n{3,}', '\n\n', text)
    return text.strip()

def split_technical(text: str, source: str) -> list[Document]:
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=950,
        chunk_overlap=140,
        separators=['\n## ', '\n### ', '\n```', '\n\n', '\n', ' ', '']
    )
    return splitter.create_documents([text], metadatas=[{'source': source, 'corpus': 'fresh_harvest_documents', 'doc_type': 'product_technical_docs'}])
````

````python
all_chunks = []
summary_rows = []

for f in files:
    rel = str(f.relative_to(REPO_ROOT))
    docs = TextLoader(str(f), encoding='utf-8').load()
    text = clean_text(docs[0].page_content if docs else '')
    if not text:
        summary_rows.append({'source': rel, 'chunks': 0, 'max_tokens': 0, 'note': 'empty_after_load'})
        continue

    chunks = split_technical(text, source=rel)
    max_tokens = 0
    for i, c in enumerate(chunks):
        t = count_tokens(c.page_content)
        c.metadata.update({'chunk_index': i, 'chunk_id': chunk_id(c.page_content), 'char_count': len(c.page_content), 'token_count': t, 'file_ext': f.suffix.lower()})
        max_tokens = max(max_tokens, t)
    all_chunks.extend(chunks)
    summary_rows.append({'source': rel, 'chunks': len(chunks), 'max_tokens': max_tokens, 'note': 'ok'})

def save_jsonl(chunks: list[Document], out_file: Path):
    with out_file.open('w', encoding='utf-8') as wf:
        for c in chunks:
            wf.write(json.dumps({'page_content': c.page_content, 'metadata': c.metadata}, ensure_ascii=False) + '\n')

save_jsonl(all_chunks, OUT_CHUNKS / 'all_fresh_harvest_documents_chunks.jsonl')
by_source = {}
for c in all_chunks:
    by_source.setdefault(c.metadata['source'], []).append(c)
for src, ch in by_source.items():
    save_jsonl(ch, OUT_CHUNKS / f"{Path(src).stem}_chunks.jsonl")

summary_df = pd.DataFrame(summary_rows)
run_id = datetime.now().strftime('%Y%m%d_%H%M%S')
summary_df.to_csv(OUT_REPORTS / f'fresh_harvest_file_summary_{run_id}.csv', index=False)
(OUT_REPORTS / f'fresh_harvest_file_summary_{run_id}.json').write_text(summary_df.to_json(orient='records', indent=2), encoding='utf-8')

print('Total chunks:', len(all_chunks))
display(summary_df.head(20))
````
